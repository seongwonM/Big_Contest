### 좌석번호 분리

```python
df['floor']=''
df['block']=''
df['seat_line']=''
df['seat_number']=0
```

```python
for index, row in df.iterrows():
    df.loc[index,'floor']=row['seat'].split()[0]
    if row['seat'].split()[1][:3]=='BOX':
        df.loc[index,'block']=row['seat'].split()[1][:4]
    else:
        df.loc[index,'block']=row['seat'].split()[1][:3]
        df.loc[index,'seat_line']=row['seat'].split()[1][3:]
    df.loc[index,'seat_number']=int(row['seat'].split()[2])
```

### 정가 찾기 - 둘다 이용

```python
df['discount_percent']=df['discount_type'].apply(extract_percentage)
df['original_price']=df['price']/(100-df['discount_percent'])*100
df[df['discount_percent'].isna()==False]['discount_percent'].astype(int)
df['original_price']=np.ceil(df['original_price'] / 1000) * 1000
df.loc[df['discount_type']=='일반','original_price']=df[df['discount_type']=='일반']['price']
```

```python
nan_indices = df[(df['original_price'].isna())|(df['original_price']==0)].index
```

```python
df_non=df[df['original_price']>0]
```

```python
for index in nan_indices:
    condition = (df_non['performance_code'] == df.loc[index, 'performance_code']) & \
                (df_non['floor'] == df.loc[index, 'floor']) & \
                (df_non['block'] == df.loc[index, 'block']) & \
                (df_non['seat_line'] == df.loc[index, 'seat_line'])
    
    rows = df_non[condition][['seat_number', 'original_price']]
    rows['key'] = rows['seat_number'] - df.loc[index, 'seat_number']
    
    if not rows['key'].empty:
        min_key_value = rows['key'].min()
        min_price = rows[rows['key'] == min_key_value]['original_price'].mode().values[0]
        df.at[index, 'original_price'] = min_price
```
