### 좌석번호 분리

```python
df['floor']=''
df['block']=''
df['seat_line']=''
df['seat_number']=0
```

```python
for index, row in df.iterrows():
    if row['seat'].split()[0]=='1층':
        df.loc[index,'floor']=1
        df.loc[index, 'block']=ord(row['seat'].split()[1][:1])-ord('A')+1
        df.loc[index, 'seat_line']=int(row['seat'].split()[1][3:-1])
    elif row['seat'].split()[0]=='2층':
        df.loc[index,'floor']=2
        if row['seat'].split()[1][:3]=='BOX':
            if int(row['seat'].split()[1][3])<=3:
                df.loc[index,'block']=int(row['seat'].split()[1][3:5].strip())
            else:
                df.loc[index,'block']=int(row['seat'].split()[1][3:5].strip())+5
        else:
            df.loc[index, 'block']=ord(row['seat'].split()[1][:1])-ord('A')+4
            df.loc[index, 'seat_line']=int(row['seat'].split()[1][3:-1])
    elif row['seat'].split()[0]=='3층':
        df.loc[index,'floor']=3
        if row['seat'].split()[1][:3]=='BOX':
            if row['seat'].split()[1][3]!='1':
                df.loc[index,'block']=int(row['seat'].split()[1][3:5].strip())-6
            else:
                df.loc[index,'block']=int(row['seat'].split()[1][3:5].strip())+1            
        else:
            df.loc[index, 'block']=ord(row['seat'].split()[1][:1])-ord('A')+4
            df.loc[index, 'seat_line']=int(row['seat'].split()[1][3:-1])      
        if df.loc[index, 'block']>=16:
            df.loc[index, 'block']=df.loc[index, 'block']-9.5
    else:
        df.loc[index,'floor']=4
        df.loc[index, 'block']=ord(row['seat'].split()[1][:1])-ord('F')+1
        df.loc[index, 'seat_line']=int(row['seat'].split()[1][3:-1])
    df.loc[index,'seat_number']=int(row['seat'].split()[2])
```

```python
df['seat_line']=df['seat_line'].replace('',0)
```

```python
df['discount_percent']=df['discount_type'].apply(extract_percentage)
df['original_price']=df['price']/(100-df['discount_percent'])*100
df[df['discount_percent'].isna()==False]['discount_percent'].astype(int)
df['original_price']=np.ceil(df['original_price'] / 1000) * 1000
df.loc[df['discount_type']=='일반','original_price']=df[df['discount_type']=='일반']['price']
```

### 좌석 찾기

```python
code=df[df['original_price']>0]['performance_code'].unique()
```

```python
code
```

```python
for x in code:
    grade=np.sort(df[(df['original_price']>0)&(df['performance_code']==x)]['original_price'].unique())
    rank_mapping = {value: rank for rank, value in enumerate(grade, start=1)}
    ranks=np.array([rank_mapping.get(number) for number in df[(df['original_price']>0)&(df['performance_code']==x)]['original_price']])
    ranks_with_nan = np.where(ranks == None, np.nan, ranks)
    ranks_with_nan = ranks_with_nan.astype(float)
    df.loc[(df['original_price']>0)&(df['performance_code']==x),'grade']=ranks_with_nan[~np.isnan(ranks_with_nan)]
```

### 좌석 채우기

```python
X=df[~df['grade'].isna()][['performance_code', 'floor', 'block', 'seat_line', 'seat_number']]
Y = df[~df['grade'].isna()]['grade']
Y = Y.astype('category')

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

X_train, Y_train = SMOTE(random_state = 0).fit_resample(X_train, Y_train)

knn_model = KNeighborsClassifier(n_neighbors = 10, metric = "euclidean")

knn_model.fit(X_train, Y_train)
Y_pred = knn_model.predict(X_test)
print("accuracy : {:.3f}".format(knn_model.score(X_train, Y_train)))
```

```python
print(classification_report(Y_test, Y_pred))
```
