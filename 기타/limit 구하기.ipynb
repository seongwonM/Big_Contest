```python
import pandas as pd
import numpy as np
import os
```

```python
df_origin=pd.read_csv('df_real_final2.csv', low_memory=False)
```

```python
df=df_origin.copy()
```

```python
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
%%time
df['limit']=np.NaN
for index, row in df.iterrows():
    if np.max(row[23:34])==row['rev_seat']:
        df.loc[index, 'limit']=row['afford']
    elif np.isnan(row.rev_seat)==False:
        df.loc[index, 'limit']=(min(row[23:34][row[23:34]>row.rev_seat])-row['min_price'])*(1-row['discount_number'])
```

```python
df.to_csv('df_real_final3.csv', index=False)
```
