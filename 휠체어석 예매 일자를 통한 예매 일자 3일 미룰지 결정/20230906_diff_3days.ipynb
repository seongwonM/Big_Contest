```python
title=pd.read_csv('seat_level.csv')
```

```python
title
```

```python
title['floor']=''
title['block']=''
title['seat_line']=''
title['seat_number']=0
```

```python
seat=title['seat']
```

```python
%%time
for index, row in enumerate(seat):
    title.at[index,'floor']=row.split()[0]
    if row.split()[1][:3]=='BOX':
        title.at[index,'block']=row.split()[1][:4]
    else:
        title.at[index,'block']=row.split()[1][:3]
        title.at[index,'seat_line']=row.split()[1][3:]
    title.at[index,'seat_number']=int(row.split()[2])
```

```python
title2=title[(title['seat'].str.contains('C블록22열'))|(title[(title['seat'].str.contains('2층 A블록8열'))])|(title[(title['seat'].str.contains('2층 E블록8열'))])]
```
